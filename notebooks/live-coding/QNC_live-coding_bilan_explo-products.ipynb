```
from pyspark.sql.functions import col, lit

columns_to_check = [
    'fat_100g',
    'saturated-fat_100g',
    'sugars_100g',
    'fiber_100g',
    'proteins_100g',
    'salt_100g',
    'sodium_100g',
    'autre',
]

filter_condition = lit(False)
for column in columns_to_check:
    condition = col(column) > 100
    filter_condition = (filter_condition | condition)

df_negative = df.filter(filter_condition)
```
